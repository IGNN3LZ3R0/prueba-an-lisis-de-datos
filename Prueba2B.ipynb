```python
import matplotlib.pylab as pl
import pandas as pd

b1 = pd.read_csv("base_starbucks.csv")
n = 50
X = np.arange(n) 
C, S = b1[' Sodium (mg)'].head(n), b1['Calories'].head(n)

pl.plot(X, C)
pl.plot(X, S)

for x, y in zip(X, C):
    pl.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')

pl.ylim(0, +350)
pl.xlim(0, +60)
pl.show()
```

```python
pl.figure(figsize=(10, 6), dpi=80)
pl.plot(X, C, color="blue", linewidth=2.5, linestyle="--")
pl.plot(X, S, color="red", linewidth=2.5, linestyle="-")
```

```python
b1 = pd.read_csv("base_starbucks.csv")
selected_categories = ["Classic Espresso Drinks", "Coffee", "Smoothies"]
filtered_data = b1[b1['Beverage_category'].isin(selected_categories)]

category_counts = filtered_data['Beverage_category'].value_counts()

labels = category_counts.index
sizes = category_counts.values
colors = ['blue', 'red', 'yellow']

plt.scatter(labels, sizes, color=colors)
plt.title('Distribución de Categorías de Bebidas')
plt.xlabel('Categoría')
plt.ylabel('Número de Elementos')

plt.show()
```

```python
b1 = pd.read_csv("base_starbucks.csv")
selected_preps = ["Grande", "Tall", "Venti"]
filtered_data = b1[b1['Beverage_prep'].isin(selected_preps)]

prep_counts = filtered_data['Beverage_prep'].value_counts()

labels = prep_counts.index
sizes = prep_counts.values
colors = ['blue', 'red', 'yellow']

plt.scatter(labels, sizes, color=colors)
plt.title('Distribución de Preparaciones de Bebidas')
plt.xlabel('Preparación')
plt.ylabel('Número de Elementos')

plt.show()
```
